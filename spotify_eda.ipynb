```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10,5)

df = pd.read_csv('/content/drive/MyDrive/spotify_tracks.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python

convert_cols = ['year', 'duration_ms', 'popularity']

for col in convert_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce').astype('Int64')

df.info()
```

```python
df['duration_min'] = df['duration_ms'] / 60000
```

```python
bins = [-1, 39, 69, 100]
labels = ['Low', 'Medium', 'High']
df['popularity_bin'] = pd.cut(df['popularity'], bins=bins, labels=labels)
```

```python
df['is_instrumental'] = df['instrumentalness'] > 0.5
```

```python
df['tempo_cat'] = pd.cut(df['tempo'], bins=[0, 90, 130, 1000], labels=['Slow', 'Medium', 'Fast'])
```

```python
print("Unique values in popularity_bin:", df['popularity_bin'].unique())
print("Unique values in tempo_cat:", df['tempo_cat'].unique())

print("\nNumber of instrumental tracks:", df['is_instrumental'].sum())
print("Total tracks:", len(df))

# Basic stats for new numeric feature
df[['duration_min','popularity']].describe()
```
